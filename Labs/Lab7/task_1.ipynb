```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../../img/letter.jpg', cv.IMREAD_GRAYSCALE)
img = cv.bitwise_not(img)
assert img is not None, "file could not be read, check with os.path.exists()"
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.imshow(img, cmap="gray")
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(img,kernel,iterations = 5)
ax2.imshow(erosion, cmap="gray")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
dilation = cv.dilate(img,kernel,iterations = 5)
ax1.imshow(img, cmap="gray")
ax2.imshow(dilation, cmap="gray")
plt.show()
```

```python
img = cv.imread('../../img/letter_noise1.jpg', cv.IMREAD_GRAYSCALE)
img = cv.bitwise_not(img)
kernel = np.ones((20,20),np.uint8)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
ax1.imshow(img, cmap="gray")
ax2.imshow(opening, cmap="gray")
plt.show()
```
