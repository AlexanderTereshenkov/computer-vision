```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('../../img/noise_logo.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"

kernel = np.ones((8,8),np.float32)/64

dst = cv.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

1. Averaging blur

```python
img = cv.imread('../../img/noise_logo.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"

blur = cv.blur(img,(5,5))

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

2. Gaussian blur

```python
img = cv.imread('../../img/noise_logo.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"

blur = cv.GaussianBlur(img,(5,5),0)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

3. Median Blurring

```python
img = cv.imread('../../img/noise_logo.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"

blur = cv.medianBlur(img,5)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

4. Bilateral Filtering

```python
img = cv.imread('../../img/carpet_2.jpeg')
assert img is not None, "file could not be read, check with os.path.exists()"

blur = cv.bilateralFilter(img,9,75,75)

fig, ax = plt.subplots(figsize=(10, 8), nrows=1, ncols=2)
ax[0].imshow(cv.cvtColor(img, cv.COLOR_BGR2RGBA))
ax[0].set_title('Original')
ax[1].imshow(cv.cvtColor(blur, cv.COLOR_BGR2RGBA))
ax[1].set_title('Blurred')

```
