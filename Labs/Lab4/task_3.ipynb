```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
im = cv.imread('../../img/polygons.png')
assert im is not None, "file could not be read, check with os.path.exists()"

imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(imgray, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
contours[0]
```

```python
img_contours = np.uint8(np.zeros((imgray.shape[0],imgray.shape[1])))

plt.imshow(cv.drawContours(img_contours, contours, -1, (255,255,255), 1))
```

```python
img_contours = np.uint8(np.zeros((imgray.shape[0],imgray.shape[1])))

plt.imshow(cv.drawContours(img_contours, contours, 60, (255,255,255), 1))
```

```python
img_contours = np.uint8(np.zeros((imgray.shape[0],imgray.shape[1])))
cnt = contours[:150]
plt.imshow(cv.drawContours(img_contours, cnt, -1, (255,255,255), 1))
```
