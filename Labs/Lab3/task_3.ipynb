```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

Resize image

```python
img = cv.imread('../../img/logo.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
 
res = cv.resize(img,None,fx=2, fy=2, interpolation = cv.INTER_CUBIC)
 
#OR
 
height, width = img.shape[:2]
res = cv.resize(img,(2*width, 2*height), interpolation = cv.INTER_CUBIC)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(cv.cvtColor(res, cv.COLOR_BGR2RGBA))
ax[1].imshow(cv.cvtColor(img, cv.COLOR_BGR2RGBA))
```

Translation

```python
img = cv.imread('../../img/logo.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

rows,cols = img.shape
 
M = np.float32([[1,0,100],[0,1,50]])
dst = cv.warpAffine(img,M,(cols,rows))
plt.imshow(dst, cmap="gray")
plt.show()
```

Rotation

```python
img = cv.imread('../../img/logo.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols = img.shape
 
# cols-1 and rows-1 are the coordinate limits.
M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),37,1)
dst = cv.warpAffine(img,M,(cols,rows))
plt.imshow(dst, cmap="gray")
plt.show()
```

Affine Transformation

```python
img = cv.imread('../../img/lines.png')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape
 
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
 
M = cv.getAffineTransform(pts1,pts2)
 
dst = cv.warpAffine(img,M,(cols,rows))
 
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

Perspective Transformation

```python
img = cv.imread('../../img/sudoku.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape
 
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
 
M = cv.getPerspectiveTransform(pts1,pts2)
 
dst = cv.warpPerspective(img,M,(300,300))
 
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```
