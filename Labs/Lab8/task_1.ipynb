```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('../../img/calibri.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img, cmap="gray")
```

```python
# cv.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]]) histSize = bins
hist = cv.calcHist([img],[0],None,[256],[0,256])
hist.shape
```

```python
# img.ravel() -> 2d array into 1d vector
hist,bins = np.histogram(img.ravel(),256,[0,256])
hist.shape, bins.shape
```

```python
plt.imshow(img, cmap="gray")
plt.show()
plt.hist(img.ravel(),100,[0,256])
plt.show()
```

```python
img2 = cv.imread('../../img/calibri.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img2],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
img3 = cv.imread('../../img/calibri.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
# create a mask
mask = np.zeros(img3.shape[:2], np.uint8)
mask[100:300, 100:400] = 255
masked_img = cv.bitwise_and(img3,img3,mask = mask)
# Calculate histogram with mask and without mask
# Check third argument for mask
hist_full = cv.calcHist([img3],[0],None,[256],[0,256])
hist_mask = cv.calcHist([img3],[0],mask,[256],[0,256])
plt.subplot(221), plt.imshow(img3, 'gray')
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img, 'gray')
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])
plt.show()
```
